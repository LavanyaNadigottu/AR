```python
# prompt: looks like there is seperator in the excel file., break it into different columns

import pandas as pd

!pip install openpyxl
df = pd.read_excel('/content/Online retail.xlsx', names=['items']) # Assuming file is in the same directory

# Assuming the separator is ';' and the column to split is 'Description'
df_split = df['items'].str.split(',', expand=True)



print(df_split.head())
```

```python
import numpy as np
dataset = df_split.replace(np.nan, '', regex=True)
dataset
```

```python
transactions = []
for i in range(0, 7500):
    # Change the range of u to 0 to 19 (exclusive) or 0 to dataset.shape[1]
    transactions.append([str(dataset.values[i,u]) for u in range(0, dataset.shape[1])])
```

```python
transactions
```

```python
groceries_series  = pd.DataFrame(pd.Series(transactions))
groceries_series
groceries_series.columns = ["business"]
groceries_series
```

```python
X = groceries_series['business'].str.join(sep=',').str.get_dummies(sep=',')
X
```
